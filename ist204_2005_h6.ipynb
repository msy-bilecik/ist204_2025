## Return Deyimi

```python
def asalmi(sayi):
  """
  Bu fonksiyon gönderilen sayının
  asal olup olmadığını döndürür.

  Args:
    sayi: integer bir sayı değeri girilir.

  Returns:
    True veya False döndürür.

  """
  for i in range(2,sayi):
    if sayi%i==0:
      return False
  return True

asalmi(5)
```

```python
# prompt: iki sayıyı bölen sonucu Bölüm ve kalan şeklinde yazan program

def bolme_islemi(bolunen, bolen):
  """
  Verilen iki sayıyı böler ve bölümü ve kalanı döndürür.

  Args:
    bolunen: Bölünecek sayı.
    bolen: Bölen sayı.

  Returns:
    Bir tuple olarak bölüm ve kalanı döndürür.
  """
  bolum = bolunen // bolen
  kalan = bolunen % bolen
  return bolum, kalan

# Kullanıcıdan girdi alma
bolunen = int(input("Bölünecek sayıyı giriniz: "))
bolen = int(input("Bölen sayıyı giriniz: "))

# Bölme işlemini yapma
bolum, kalan = bolme_islemi(bolunen, bolen)

# Sonuçları yazdırma
print("Bölüm:", bolum)
print("Kalan:", kalan)
```

```python
bolum,_=bolme_islemi(89,11)
print(bolum)
```

# Öz Yinelemeli (Recursive) Fonksiyonlar

Not: Bir fonksiyonun öz yinelemeli olması için 2 temel şart vardır.

*   Kendi kendini çağırması
*   Durma Noktası

```python
def fakt(n):  # normal faktöriyel fonksiyonu
  f=1
  for i in range(1,n+1):
    f*=i
  return f

def faktR(n):
  if n==1:
    return 1
  else:
    return n*faktR(n-1)
```

```python
print(fakt(50))
print(faktR(50))
```
